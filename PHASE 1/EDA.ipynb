```python
sns.boxplot(x=df['area_sqft'])
plt.title("Before Outlier Removal")
plt.show()
```

```python
Q1 = df['area_sqft'].quantile(0.25)
Q3 = df['area_sqft'].quantile(0.75)

# Interquartile Range (IQR)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['area_sqft'] < lower_bound) | (df['area_sqft'] > upper_bound)]
print("Outliers identified:")
print(outliers)

# outliers removed
df= df[(df['area_sqft'] >= lower_bound) & (df['area_sqft'] <= upper_bound)]
```

```python
print("Number of outliers:", len(outliers))
```

```python
df= df[(df['area_sqft'] >= lower_bound) & (df['area_sqft'] <= upper_bound)]
```

```python
Q1 = df['area_sqft'].quantile(0.25)
Q3 = df['area_sqft'].quantile(0.75)

# Interquartile Range (IQR)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['area_sqft'] < lower_bound) | (df['area_sqft'] > upper_bound)]
print("Outliers identified:")
print("Number of outliers:", len(outliers))
```
